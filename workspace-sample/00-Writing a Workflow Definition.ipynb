# Writting a Workflow Definition

```python
import sys
import asyncio
```

### Business Logic

```python
class GreetSomeone:
    async def run(self, name: str) -> str:
        return f"Hello {name}!"
```

```python
name = input("Digite o nome:")
greeter = GreetSomeone()
greeting = await greeter.run(name)
print(greeting)
```

## Workflow Definition

```python
from temporalio import workflow


@workflow.defn
class GreetSomeone:
    @workflow.run
    async def run(self, name: str) -> str:
        return f"Hello {name}!"
```
